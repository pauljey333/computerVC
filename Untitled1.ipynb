Image Addition, the sizes of the pictures should be the same. To be more precise the sizes of the matrix.

```python
img1 = cv2.imread(filename='bear.jpg')
img2 = cv2.imread(filename='view.jpg')
add_img = cv2.add(src1=img1, src2=img2)
add_img.shape
cv2_imshow(add_img)
```

Blendowanie

```python
dst = cv2.addWeighted(src1=img1, alpha=0.3, src2=img2, beta=0.7, gamma=0)
cv2_imshow(dst)
```
